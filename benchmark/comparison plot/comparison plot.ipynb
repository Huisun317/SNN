```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df02=pd.read_csv('02_cnnTest.csv')
df2=pd.read_csv('2_cnnTest.csv')
df4=pd.read_csv('4_cnnTest.csv')
```

```python
df02_snn=pd.read_csv('02_snnTest.csv')
df2_snn=pd.read_csv('2_snnTest.csv')
df4_snn=pd.read_csv('4_snnTest.csv')
```

```python
df_cnnTest=pd.DataFrame()
```

```python
df_cnnTest['2e-2 Epsilon CNN']=df02['0']
df_cnnTest['2 Epsilon CNN']=df2['0']
df_cnnTest['4 Epsilon CNN']=df4['0']

df_cnnTest['2e-2 Epsilon SNN']=df02_snn['0']
df_cnnTest['2 Epsilon SNN']=df2_snn['0']
df_cnnTest['4 Epsilon SNN']=df4_snn['0']
```

```python
df_cnnTest[['2e-2 Epsilon CNN', '2e-2 Epsilon SNN']].plot(figsize=(8,6), style=['-','-.'],color=['blue', 'red'])
plt.legend(['$\epsilon=0.2$ CNN','$\epsilon=0.2$ SNN'],fontsize=15)
plt.grid('-.')
plt.xlabel('Epochs')
plt.ylabel(' Test Accuracy ')
plt.title('Comparison between noisy CNN and SNN Test accuracy ')
plt.savefig('test02.png')
```

```python
df_cnnTest[['2 Epsilon CNN', '2 Epsilon SNN']].plot(figsize=(8,6), style=['-','-.'],color=['blue', 'red'])
plt.legend(['$\epsilon=2.0$ CNN','$\epsilon=2.0$ SNN'],fontsize=15)
plt.grid('-.')
plt.xlabel('Epochs')
plt.ylabel(' Test Accuracy ')
plt.title('Comparison between noisy CNN and SNN Test accuracy ')
plt.savefig('test2.png')
```

```python
df_cnnTest[['4 Epsilon CNN', '4 Epsilon SNN']].plot(figsize=(8,6), style=['-','-.'],color=['blue', 'red'])
plt.legend(['$\epsilon=4.0$ CNN','$\epsilon=4.0$ SNN'],fontsize=15)
plt.grid('-.')
plt.xlabel('Epochs')
plt.ylabel(' Test Accuracy ')
plt.title('Comparison between noisy CNN and SNN Test accuracy ')
plt.savefig('test4.png')
```
